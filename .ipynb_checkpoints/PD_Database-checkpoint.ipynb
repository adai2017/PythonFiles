```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy.random as np
import sys
import matplotlib
import re
%matplotlib inline
```

```python
df = pd.read_excel('PD_workbook_mk2.xlsx', sheetname=None, header=None)
data = df["Sheet1"]
data
```

```python
Participants = data[:1]
ParticipantStatus = data[1:2]
NextScheduledVisit = data[2:3]
RegisteredWithNMH = data[6:7]
AddedToApp = data[7:8]
CompletedDay1 = data[8:9]
CompletedDay2 = data[9:10]
DataReady = data[10:11]
DataProcessed = data[11:12]

d = {}
d["1"] = "Not Registered with NMH"
d["2"] = 'Registered with NMH'
d["3"] = 'Added to PD study app'
d["4"] = 'Completed Day 1 Testing'
d["5"] = 'Completed Day 2 Testing'
d["6"] = 'Data Ready for Processing'
d["7"] = 'Data Processed'

e = {}
e["1004"] = "1"
e["1005"] = "2"
e["1006"] = "3"
e["1007"] = "4"
e["1008"] = "5"
e["1009"] = "6"
e["PD_07"] = "7"
e["PD_08"] = "8"
e["PD_09"] = "9"
e["PD_10"] = "10"
e["PD_11"] = "11"
e["PD_12"] = "12"
e["PD_13"] = "13"
e["PD_14"] = "14"
e["PD_15"] = "15"

f = {}
f["01"] = "January"
f["02"] = "February"
f["03"] = "March"
f["04"] = "April"
f["05"] = "May"
f["06"] = "June"
f["07"] = "July"
f["08"] = "August"
f["09"] = "September"
f["10"] = "October"
f["11"] = "November"
f["12"] = "December"

ParticipantOfInterest = input("Enter 4 digit participant number: ")
DesiredData = str(input("Press 1 for Current Status \nPress 2 for Next Scheduled Visit \nPress 3 for Status History \nValue Entered: "))

DataLocation = e[ParticipantOfInterest]
DataLocationInt = int(DataLocation)

StatusNumber = ParticipantStatus.loc[1,DataLocationInt]
StatusDescriptionLocation = str(StatusNumber)
StatusDescription = d[StatusDescriptionLocation]

DateNumberTot = NextScheduledVisit.loc[2,DataLocationInt]
DateNumberTotStr = str(DateNumberTot)

DateHistory1 = str(RegisteredWithNMH.loc[6,DataLocationInt])
DateHistory2 = str(AddedToApp.loc[7,DataLocationInt])
DateHistory3 = str(CompletedDay1.loc[8,DataLocationInt])
DateHistory4 = str(CompletedDay2.loc[9,DataLocationInt])
DateHistory5 = str(DataReady.loc[10,DataLocationInt])
DateHistory6 = str(DataProcessed.loc[11,DataLocationInt])
     

def excel_date_format(DateNumber):
    DateNumberStr = str(DateNumber)
    Year = DateNumberStr[:4]
    MonthNumber = DateNumberStr[5:7]
    MonthStr = str(MonthNumber)
    Month = f[MonthStr]
    Day = DateNumberStr[8:10]
    DateStr = Month+" "+Day+", "+Year
    return DateStr


if DesiredData == '1':
    print ('\nThe current status of Participant',ParticipantOfInterest,'is level',StatusNumber,'-',StatusDescription) 
elif DesiredData == '2':
    DateNumberTotCheck = str(DateNumberTot)
    if DateNumberTotCheck[0] == 'n':
        print ('\nParticipant',ParticipantOfInterest,'is not currently scheduled for another visit')
    else: 
        DateStr = excel_date_format(DateNumberTot)
        print ('\nParticipant',ParticipantOfInterest,'is scheduled for a visit on',DateStr)
elif DesiredData == '3':
    if DateHistory1[0] == 'n':
        print ("\nParticipant",ParticipantOfInterest,"has not yet completed any study objectives")
    elif DateHistory2[0] == 'n':
        DateStr = excel_date_format(DateHistory1)
        print ("\nParticipant",ParticipantOfInterest,"was registered with NMH on:",DateStr)
    elif DateHistory3[0] == 'n':
        DateStr1 = excel_date_format(DateHistory1)
        DateStr2 = excel_date_format(DateHistory2)
        print ("\nParticipant",ParticipantOfInterest,"was registered with NMH on:",DateStr1,"\n\nParticipant",ParticipantOfInterest,"was added to the PD study app on:",DateStr2)
    elif DateHistory4[0] == 'n':
        DateStr1 = excel_date_format(DateHistory1)
        DateStr2 = excel_date_format(DateHistory2)
        DateStr3 = excel_date_format(DateHistory3)
        print ("\nParticipant",ParticipantOfInterest,"was registered with NMH on:",DateStr1,"\n\nParticipant",ParticipantOfInterest,"was added to the PD study app on:",DateStr2,"\n\nParticipant",ParticipantOfInterest,"completed dat 1 testing on:",DateStr3)
    elif DateHistory5[0] == 'n':
        DateStr1 = excel_date_format(DateHistory1)
        DateStr2 = excel_date_format(DateHistory2)
        DateStr3 = excel_date_format(DateHistory3)
        DateStr4 = excel_date_format(DateHistory4)
        print ("\nParticipant",ParticipantOfInterest,"was registered with NMH on:",DateStr1,"\n\nParticipant",ParticipantOfInterest,"was added to the PD study app on:",DateStr2,"\n\nParticipant",ParticipantOfInterest,"completed dat 1 testing on:",DateStr3,"\n\nParticipant",ParticipantOfInterest,"completed day 2 testing on:",DateStr4)
    elif DateHistory6[0] == 'n':
        DateStr1 = excel_date_format(DateHistory1)
        DateStr2 = excel_date_format(DateHistory2)
        DateStr3 = excel_date_format(DateHistory3)
        DateStr4 = excel_date_format(DateHistory4)
        DateStr5 = excel_date_format(DateHistory5)
        print ("\nParticipant",ParticipantOfInterest,"was registered with NMH on:",DateStr1,"\n\nParticipant",ParticipantOfInterest,"was added to the PD study app on:",DateStr2,"\n\nParticipant",ParticipantOfInterest,"completed dat 1 testing on:",DateStr3,"\n\nParticipant",ParticipantOfInterest,"completed day 2 testing on:",DateStr4,"\n\nData for participant",ParticipantOfInterest," has been ready for processing since:",DateStr5)
    elif DateHistory6[0] != 'n':
        DateStr1 = excel_date_format(DateHistory1)
        DateStr2 = excel_date_format(DateHistory2)
        DateStr3 = excel_date_format(DateHistory3)
        DateStr4 = excel_date_format(DateHistory4)
        DateStr5 = excel_date_format(DateHistory5)
        DateStr6 = excel_date_format(DateHistory6)
        print ("\nParticipant",ParticipantOfInterest,"was registered with NMH on:",DateStr1,"\n\nParticipant",ParticipantOfInterest,"was added to the PD study app on:",DateStr2,"\n\nParticipant",ParticipantOfInterest,"completed dat 1 testing on:",DateStr3,"\n\nParticipant",ParticipantOfInterest,"completed day 2 testing on:",DateStr4,"\n\nData for participant",ParticipantOfInterest,"was ready for processing on:",DateStr5,"\n\nData for participant",ParticipantOfInterest,"was processed on:",DateStr6)
```
