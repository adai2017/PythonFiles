```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

plt.rcParams['figure.figsize'] = (15, 5)
plt.style.use('ggplot')
```

```python
Location = r'C:\Users\adai\Documents\GitHub\pandas-cookbook\data\bikes.csv'
bikes = pd.read_csv(Location, sep=';', encoding='latin1', parse_dates=['Date'], dayfirst=True, index_col='Date')
bikes['Berri 1'].plot()
```

```python
berri_bikes = bikes[['Berri 1']]
berri_bikes.head()
```

```python
berri_bikes['weekday'] = berri_bikes.index.weekday
berri_bikes.head()
```

```python
weekday_counts = berri_bikes.groupby('weekday').aggregate(sum)
weekday_counts
```

```python
weekday_counts.index = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
weekday_counts
```

```python
weekday_counts.plot(kind='bar')
```

```python
bikes = pd.read_csv(Location, sep=';', encoding='latin1', parse_dates=['Date'], dayfirst=True, index_col='Date')

berri_bikes = bikes[['Berri 1']]
berri_bikes['weekday'] = berri_bikes.index.weekday

weekday_counts = berri_bikes.groupby('weekday').aggregate(sum)
weekday_counts.index = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
weekday_counts.plot(kind='bar')
```
